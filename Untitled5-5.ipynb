```python
import pandas as pd

df = pd.read_csv('Chicago_Crimes_2012_to_2017.csv')

df.head()
```

```python
missing_values = df.isnull().sum()

missing_values
```

```python
df_cleaned = df.dropna()

rows_after_cleaning = df_cleaned.shape[0]

rows_dropped = df.shape[0] - rows_after_cleaning
rows_dropped, rows_after_cleaning
```

```python
from sklearn.model_selection import train_test_split

features = ['Latitude', 'Longitude', 'District', 'Ward', 'Domestic', 'Distance_from_center']
target = 'Arrest' 

X = df_cleaned[features]
y = df_cleaned[target]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

X_train.shape, X_test.shape
```

```python
X.head()
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report

le_district = LabelEncoder()
le_ward = LabelEncoder()
le_arrest = LabelEncoder()
le_domestic = LabelEncoder()

X_train['District'] = le_district.fit_transform(X_train['District'])
X_train['Ward'] = le_ward.fit_transform(X_train['Ward'])
X_train['Domestic'] = le_domestic.fit_transform(X_train['Domestic'])

X_test['District'] = le_district.transform(X_test['District'])
X_test['Ward'] = le_ward.transform(X_test['Ward'])
X_test['Domestic'] = le_domestic.transform(X_test['Domestic'])

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

clf = DecisionTreeClassifier(max_depth=10, random_state=42)
clf.fit(X_train_scaled, y_train)

y_pred = clf.predict(X_test_scaled)

accuracy = accuracy_score(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

accuracy, classification_rep
```

```python
from sklearn.utils import resample

# Specify the desired ratio (e.g., 1:1)
desired_ratio = 1.0

# Calculate the number of majority class samples to keep
minority_class_count = sum(y_train)
majority_class_count = len(y_train) - minority_class_count
undersample_ratio = minority_class_count / majority_class_count

# Perform random undersampling
if undersample_ratio < desired_ratio:
    # If the current ratio is less than desired, perform undersampling
    majority_class_indices = y_train[y_train == 0].index
    undersampled_indices = resample(majority_class_indices, n_samples=int(majority_class_count * desired_ratio),
                                    random_state=42)
    X_undersampled = pd.concat([X_train.loc[undersampled_indices], X_train.loc[y_train == 1]])
    y_undersampled = pd.concat([y_train.loc[undersampled_indices], y_train.loc[y_train == 1]])
else:
    # If the current ratio is equal to or greater than desired, no undersampling is needed
    X_undersampled = X_train
    y_undersampled = y_train
```

```python
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split

# Step 2: Train and evaluate the model on the original high-dimensional data
# Create and train a Random Forest classifier on the original data
rf_model_original = RandomForestClassifier(random_state=42)
rf_model_original.fit(X_train, y_train)

# Make predictions and evaluate the model
y_pred_original = rf_model_original.predict(X_test)
accuracy_original = accuracy_score(y_test, y_pred_original)
report_original = classification_report(y_test, y_pred_original)

# Step 3: Apply dimensionality reduction (e.g., PCA)
n_components = 50  # Choose the number of components for PCA
pca = PCA(n_components=n_components)

X_train_reduced = pca.fit_transform(X_train)
X_test_reduced = pca.transform(X_test)

# Step 4: Train and evaluate the model on the reduced-dimension data
# Create and train a Random Forest classifier on the reduced-dimension data
rf_model_reduced = RandomForestClassifier(random_state=42)
rf_model_reduced.fit(X_train_reduced, y_train)

# Make predictions and evaluate the model
y_pred_reduced = rf_model_reduced.predict(X_test_reduced)
accuracy_reduced = accuracy_score(y_test, y_pred_reduced)
report_reduced = classification_report(y_test, y_pred_reduced)

# Compare performance
print("Performance on Original Data:")
print("Accuracy:", accuracy_original)
print("Classification Report:\n", report_original)

print("\nPerformance on Reduced-Dimension Data:")
print("Accuracy:", accuracy_reduced)
print("Classification Report:\n", report_reduced)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

# Define candidate models and their respective hyperparameter grids
models = [
    ('RandomForest', RandomForestClassifier(), {'n_estimators': [50, 100, 200], 'max_depth': [None, 10, 20]}),
    ('XGBoost', XGBClassifier(), {'n_estimators': [50, 100, 200], 'max_depth': [3, 5, 7], 'learning_rate': [0.1, 0.01]})
]

# Perform grid search with cross-validation for each model
best_model = None
best_score = 0

for model_name, model, param_grid in models:
    grid_search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy')
    grid_search.fit(X_train, y_train)
    
    # Get the best hyperparameters and CV score for this model
    best_params = grid_search.best_params_
    cv_score = grid_search.best_score_
    
    print(f"Model: {model_name}")
    print(f"Best Hyperparameters: {best_params}")
    print(f"CV Score (Accuracy): {cv_score:.4f}")
    
    # Check if this model outperforms the current best model
    if cv_score > best_score:
        best_score = cv_score
        best_model = grid_search.best_estimator_

# Train the best model on the full training dataset
best_model.fit(X_train, y_train)

# Evaluate the best model on a holdout test set
test_score = best_model.score(X_test, y_test)
print(f"Test Score (Accuracy): {test_score:.4f}")
```
